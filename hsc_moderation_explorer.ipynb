# HSC Moderation explorer

This simulation lets you explore the effect of HSC moderation on student scores by varying a range of input parameters. The model generates random cohorts of school assessmnet scores and HSC examination marks according to the specified input parameters. 
 - school_size: the number of sudents in the subject at the school
 - school_mean: the mean of the school assessment scores
 - school_sd: the standard deviation of the school assessment scores
 - school_max: the maximum school mark (this is restricted to values between 2 and 98 in this simulation)
 - delta_mean: the mean of the difference between school assessment and hsc scores. Set a positive number if you expect hsc exam scores to be greater than school assessment scores, and a negative number if you expect the exam scores will be lower. Default is 0.
 - delta_sd: the standard deviation of the difference between school assessment and hsc scores.
 - tied: a string indicating whether you want to analyse the effect of splitting the top school scores if they are tied. The valid values are 'tie' to specify class scores with at least two students are tied; 'pair' specify exactly 2 tied scores; 'triple' to specify exactly 3 tied score; 'none' to specify 0 tied scores; and, 'normal' no restriction on tied scores.
 - runs: the number of data sets or iterations to compare

The default values for these imputs are: school_size = 15, school_mean = 75, school_sd = 13, school_max = 95, delta_mean = 0, delta_sd = 5, tied = 'tied', runs = 1000.

After changing some parameters, run the sim by pressing the black arrow at the top left of the code window below

#### Note: if you choose to specify 'tie', 'pair' or 'triple' the model will compare the outcome of moderation of the tied scores to moderation of the scores after splitting the tie by increasing the score of 1 student.

```python
# explore the effects of moderation by changing the values below
school_size = 10
school_mean = 75
school_sd = 13
school_max = 95
delta_mean = 0
delta_sd = 5
runs = 1000
tied = 'tie'

#### WARNING! ATTENTION! ACHTUNG! dont change stuff below this line ####
import hsc_helpers2 as hh
school_max = min(98, school_max)
school_size = max(2, school_size)
input_params = [school_size, school_mean, school_sd, school_max, delta_mean, delta_sd, runs, tied]
results = hh.multi_run(input_params)
hh.multi_run_plot(results, input_params)
```
